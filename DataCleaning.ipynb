```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
```

DATA CLEANING STARTS


```python
##LOOK FOR MISSING VALUES
df[df.isnull().any(axis=1)]##axis 1 means rows and axis 0 means cols
```

```python
df.loc[:122,"region"]=0
df.loc[122:,"region"]=1
```

```python
df.info()
```

```python
##convertion of region float into int
```

```python
df[['region']] = df[["region"]].astype(int)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
##fixing spaces in columns name
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtype=='o']
```

####Generic method to convert data types of many cols togetherðŸ‘†ðŸ‘‡


```python
for i in objects:
    if i != 'Classes':               ##becoz class is our output feature
        df[i] = df[i].astype(float)
```
