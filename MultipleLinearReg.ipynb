```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_index = pd.read_csv('economic_index.csv')
df_index.head()
```

```python
df_index.drop(columns=['Unnamed: 0','year','month'],inplace=True)##removing unneccesary columns
```

```python
df_index.head()
```

```python
df_index.isnull().sum()
```

```python
df_index.corr()
```

```python
sns.pairplot(df_index)
```

```python
plt.scatter(df_index['interest_rate'], df_index['unemployment_rate'])
```

```python
X=df_index.iloc[:,:-1]
y=df_index.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
validation=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3)
np.mean(validation)
```

```python
y_pred = regression.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mae,mse,rmse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
